```python
# 1. First step: Load data into numpy, pandas or both
import pandas as pd
data = pd.read_csv(r'C:\Users\sweet\Desktop\python_boot_camp\iris.csv')
#data.columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'class']


# 2. print head and tail of data
h = data.head()
t = data.tail()
print(f"The first 5 rows of iris dataset:\n{h}\n")
print(f"The last 5 rows of iris dataset:\n{t}\n")


# 3. Clean the data if needed
missing = data.apply(lambda x: sum(x.isnull()), axis=0) 
print(f"The miising values by columns:\n{missing}\n")
print(f"Iris data shape:\n{data.shape}\n")


# 4. import the package Matplotlib, Sklean\
# 5. Visulaize the data in different types of figures as you see fit
# 6. Be creative!! and use clustering or classification methods on the data

import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# features
x = data.iloc[:, :-1].values

# y target (label)
y = data.iloc[:, 4].values


# data clustering using Kmeans 
kmeans = KMeans(n_clusters=3)
kmeans.fit(x)
y_predict = kmeans.predict(x)
print(y_predict)


# clusters visualization
plt.scatter(x[:, 0], x[:, 1], c=y_predict, s=50, cmap='viridis')
plt.show()

```
