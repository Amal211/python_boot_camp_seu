# Numpy Practice

```python
# 1. import the package numpy and make the shortcut "np" for it.
import numpy as np


# 2. define a numpy array [1, 2, 3, 4, 5] and assign it to variable x
x = np.array([1, 2, 3, 4, 5])
print(f"A numpy array of 5 integer values:\n{x}\n")


# 3. Print the mean of x using numpy
print(f"The mean value of x:\n{np.mean(x)}\n")


# 4. Print the maximum number in array x
print(f"The maximum value of x:\n{np.max(x)}\n")


# 5. Generate a random number between 35 and 40
y = np.random.randint(low = 35, high = 41)
print(f"Random number between 35 and 40:\n{y}\n")


# 6. generate a list (N) of 1000 numbers with random values between 100 and 150
n = np.random.randint(low = 100, high = 151, size = 1000)
print(f"List of 1000 numbers with values between 100 and 150:\n{n}\n")


# 7. get the mean, median, std, and variance of the list N
print(f"The mean value of the list n:\n{np.mean(n)}\n")
print(f"The median value of the list n:\n{np.median(n)}\n")
print(f"The std value of the list n:\n{np.std(n)}\n")
print(f"The variance value of the list n:\n{np.var(n)}\n")


# 8. generate a 2D array of size 3 x 4 of random numbers between 18-65
arr = np.random.randint(low = 18, high = 66, size = (3, 4))
print(f"2D array of size 3 Ã— 4 of random numbers between 18 and 65:\n{arr}\n")


# 9. Get the mean for every columns N
print(f"The mean value for every column in the 2D array:\n{np.mean((arr), axis = 1)}\n")


# 10. Get the mean for every row of N
print(f"The mean value for every row in the 2D array:\n{np.mean((arr), axis = 0)}\n")


# 11. Create the following 2D array (S) of students grade in a classroom
# grades [coursework, midterm exam, final exam]
# [20, 15, 40]
# [25, 24, 35]
# [21, 15, 25]
s = np.array([[20, 15, 40], [25, 24, 35], [21, 15, 25]])
print(f"2D array of students grade in a classroom:\n{s}\n")


# 12. Get the sums of student grades in S
print(f"The sum of student grades:\n{np.sum((s), axis = 1)}\n")


# 13. Get the means of student grades in every activity 
# (coursework, midterm exam, final exam) in S
print(f"The means of student grades in every activity:\n{np.mean((s), axis = 0)}\n")


# 14. reshape S in a way that make the colums represent the students and rows
# represent the activity, as follow
# [20, 25, 21]
# [15, 24, 15]
# [40, 35, 25]
arr_s = np.swapaxes(s,0,1)
print(f"Reshape the s array:\n{(arr_s)}\n")

```

# Pandas Practice

```python
# 1. import the package pandas and make the shortcut "pd" for it.
import pandas as pd


# 2. create Series call it (Data) with the values 'a','b','d','u' and 'l'
d = ["a", "b", "d", "u", "l"]
data = pd.Series(d)
print(f"Simple Pandas series:\n{data}\n")


# 3. change index to 3,6,9,12, and 15 for (data)
data = data.reindex([3, 6, 9, 12, 15])
print(f"Simple Pandas series with the new index:\n{data}\n")


# 4. change index to client1, client2, client3, client4 and client5 for (data)
data = data.reindex(['client1', 'client2', 'client3', 'client4', 'client5'])
print(f"Simple Pandas series with the 2nd new index:\n{data}\n")


# 5. create a dataframe (D) with the followin values
# Name - Age
# Ali - 55
# Ahmad - 15
# Nora - 40
dict = {
    "Name" : ["Ali", "Ahmad", "Nora"],
    "Age" : [55, 15, 40]
       } 
d = pd.DataFrame(dict)
print(f"Simple Pandas dataframe:\n{d}\n")


# 6. Get the mean of ages of people in (D)
print(f"The mean of ages of people in D:\n{d['Age'].mean()}\n")

```
